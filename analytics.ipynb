```python
import json
import matplotlib.pyplot as plt
import numpy as np
import glob
import pandas as pd
import seaborn as sns

path = "outputs/"
algo_names = ["1nn", "cart", "mlp", "nb", "svm", "random_forest"]
def load_results_from_term(term):
    results_data = {}
    for algo_name in algo_names:
        path_full = glob.glob(path + "full_" + term + "_" + algo_name + "*")[0]
        path_reduced = glob.glob(path + "reduced_" + term + "_" + algo_name + "*")[0]
        path_resampled = glob.glob(path + "resampled_" + term + "_" + algo_name + "*")[0]
        path_improved = glob.glob(path + "improved_" + term + "_" + algo_name + "*")[0]
        print(f"Full {algo_name}:  {path_full}")
        print(f"Reduced {algo_name}:  {path_reduced}")
        print(f"Resampled {algo_name}:  {path_resampled}")
        print(f"Improved {algo_name}:  {path_improved}")
        results_data[algo_name] = {
            "full": json.load(open(path_full)),
            "reduced": json.load(open(path_reduced)),
            "resampled": json.load(open(path_resampled)),
            "improved": json.load(open(path_improved)) 
        }
    
    return results_data



arit = load_results_from_term("arit")
print("-----------------------")
esc = load_results_from_term("esc")
print("---------------------------")
leit = load_results_from_term("leit")
```

```python
import hashlib

def create_counter_df(df, label_name):
    data = []
    for i, row in df.iterrows():
        str_row = str(row.to_list())
        hash_value = hashlib.sha256(str_row.encode('utf-8')).hexdigest()
        label = row[label_name]
        data.append([i, label, hash_value, 0])

    return pd.DataFrame(data=data, columns=["instance", "label", "hash", "freq"])

df_arit_counter = create_counter_df(arit_train_dataset, "TDE_MG_Arit")
df_esc_counter = create_counter_df(esc_train_dataset, "TDE_MG_Esc")
df_leit_counter = create_counter_df(leit_train_dataset, "TDE_MG_Leit")
df_leit_counter
# x = df_arit_counter.loc[df_arit_counter['hash'] == "6c9abdab20c6724493d948ae60ab733d2826f4aef387eb188ef00d72bee"]
```

```python
def count_selected_instance_freq(database, counter_df, improved=True):
    df = counter_df.copy()
    results = []
    
    if improved:
        term = "improved"
    else:
        term = "reduced"
    
    for algo_name in algo_names:
        results.append(database[algo_name][term])

    for result in results:
        for key in result:
            if key.startswith('i'):
                selected_instances_in_cv = result[key]["selected_indices"]
                for selected_instances in selected_instances_in_cv:
                    for selected_instance in selected_instances:
                        df.loc[df['hash'] == selected_instance, 'freq'] += 1
        
    return df.sort_values(by='freq', ascending=False)

                   

selected_df_arit = count_selected_instance_freq(arit, df_arit_counter)
selected_df_esc = count_selected_instance_freq(esc, df_esc_counter)
selected_df_leit = count_selected_instance_freq(leit, df_leit_counter)
```

```python
# get mean results obianted in all results
def get_class_balance(data, label_result):
    results = []
    for algo_name in algo_names:
        results.append(data[algo_name][label_result])
    
    sup_ratio_ori = []
    inf_ratio_ori = []
    sup_ratio_red = []
    inf_ratio_red = []
    sup_ratio_res = []
    inf_ratio_res = []
    for result in results:
        for result in results:
            for key in result:
                if key.startswith('i'):
                    for ratios in result[key]['class_info']:
                        sup_ratio_ori.append(ratios["sup_rate_original"])
                        inf_ratio_ori.append(ratios["inf_rate_original"])
                        sup_ratio_red.append(ratios["sup_rate_reduced"])
                        inf_ratio_red.append(ratios["inf_rate_reduced"])
                        sup_ratio_res.append(ratios["sup_rate_resampled"])
                        inf_ratio_res.append(ratios["inf_rate_resampled"])
    return {
        "sup_rate_ori": np.mean(sup_ratio_ori),
        "inf_rate_ori": np.mean(inf_ratio_ori),
        "sup_ratio_red": np.mean(sup_ratio_red),
        "inf_ratio_red": np.mean(inf_ratio_red),
        "sup_ratio_res": np.mean(sup_ratio_res),
        "inf_ratio_res": np.mean(inf_ratio_res)
    }

print(get_class_balance(arit, "improved"))
print()
print(get_class_balance(esc, "improved"))
print()
print(get_class_balance(leit, "improved"))
```

```python
# getting info of number of instances
def get_total_instances(data, label_result):
    results = []
    for algo_name in algo_names:
        results.append(data[algo_name][label_result])
    
    total_instances_ori = []
    total_instances_red = []
    total_instances_res = []
    for result in results:
        for result in results:
            for key in result:
                if key.startswith('i'):
                    total_instances_ori += result[key]['total_instances_ori']
                    total_instances_red += result[key]['total_instances_red']
                    total_instances_res += result[key]['total_instances_res']
    return {
        "total_instances_ori": np.mean(total_instances_ori),
        "total_instances_red": np.mean(total_instances_red),
        "total_instances_res": np.mean(total_instances_res)
    }

print(get_total_instances(arit, "improved"))
print()
print(get_total_instances(esc, "improved"))
print()
print(get_total_instances(leit, "improved"))
```
